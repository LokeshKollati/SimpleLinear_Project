```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("Salary_dataset.csv")
```

```python
df.drop("Unnamed: 0",inplace=True,axis=1)
```

```python
sns.scatterplot(x='YearsExperience',y='Salary',data=df)
plt.scatter( x=df['YearsExperience'],y=df['Salary'], c='blue')
plt.show()
```

```python
df.head()
```

```python
x=df[['YearsExperience']]
np.array(x).ndim
y=df['Salary']
np.array(y).ndim
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
# x_train.sort_values(by='YearsExperience').info()
```

```python
(0.5*2.67)+4.89
```

```python
from sklearn.preprocessing import StandardScaler
Scalar = StandardScaler()
x_train=Scalar.fit_transform(x_train)
# print(x_test)
x_test=Scalar.transform(x_test)
# print(x_test)
# np.std(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(x_train,y_train)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,model.predict(x_train))
```

```python
model.predict(Scalar.transform([[6.225]]))
```

```python
y_pred=model.predict(x_test)
plt.scatter(x=x_test,y=y_test,label='dataset')
plt.scatter(x=x_test,y=y_pred,label='model')
plt.legend()
```

```python
# plt.plot(x_test,y_test,label='dataset',linestyle='--')
plt.scatter(x=x_test,y=y_test,label='dataset')
plt.plot(x_test,y_pred,label='model',color='red')
plt.legend()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print("mse:",mean_squared_error(y_test,y_pred))
print("r2score",r2_score(y_test,y_pred))
```
